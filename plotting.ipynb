```python
from ScreeningFactorData import ScreeningFactorData, ExpLogData, MassFractionData, pyna
import numpy as np

rng = np.random.default_rng
```

```python
nuclei = ["h1", "he4", "c12", "o16", "n14", "ca40"]
num_nuclei = len(nuclei)
comp = pyna.Composition([pyna.Nucleus(n) for n in nuclei])

reactants = ["c12", "he4"]

temp_range = (1.e7, 1.e10)
dens_range = (1.e4, 1.e8)

size = 10**3

data_kwargs = {
    "comp": comp,
    "reactants": reactants,
    "threshold": 1.01
}

_x = np.linspace(0, 1, size)
```

```python
test = ScreeningFactorData(
    temperatures = ExpLogData.from_uniform(data_range = temp_range, scaled = _x, size = size, rng = rng(6)),
    densities = ExpLogData.from_uniform(data_range = dens_range, scaled = _x, size = size, rng = rng(7)),
    mass_fractions = MassFractionData.from_static([1/6]*6, size = size),
    **data_kwargs
)
```

```python
np.max(test.factors)
```
