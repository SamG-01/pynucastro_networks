```python
from ScreeningFactorData import ScreeningFactorData, ExpLogData, MassFractionData, pyna
import numpy as np

rng = np.random.default_rng
```

```python
nuclei = ["h1", "he4", "c12", "o16", "n14", "ca40"]
num_nuclei = len(nuclei)
comp = pyna.Composition([pyna.Nucleus(n) for n in nuclei])

reactants = ["c12", "he4"]

temp_range = (1.e7, 7.e8)
dens_range = (1.e4, 1.e8)

size = 10**3

data_kwargs = {
    "comp": comp,
    "reactants": reactants,
    "threshold": 1.01
}

_x = np.linspace(0, 1, size)

uniform_fractions = MassFractionData.from_static([1/6]*6, size = size)
```

```python
test = ScreeningFactorData(
    temperatures = ExpLogData.from_uniform(data_range = temp_range, scaled = _x, size = size, rng = rng(6)),
    densities = ExpLogData.from_uniform(data_range = dens_range, scaled = _x, size = size, rng = rng(7)),
    mass_fractions = uniform_fractions,
    **data_kwargs
)
test.plot_factors()
```

```python
test2 = ScreeningFactorData(
    temperatures = ExpLogData.from_uniform(data_range = (1.e7, 2.e7), scaled = _x, size = size, rng = rng(6)),
    densities = ExpLogData.from_uniform(data_range = (1.e4, 8.e4), scaled = _x, size = size, rng = rng(7)),
    mass_fractions = uniform_fractions,
    **data_kwargs
)
test2.plot_factors()
```

```python
test3 = ScreeningFactorData(
    temperatures = ExpLogData.from_uniform(data_range = temp_range, scaled = _x, size = size, rng = rng(6)),
    densities = ExpLogData.from_uniform(data_range = dens_range, scaled = _x, size = size, rng = rng(7)),
    mass_fractions = MassFractionData.from_static([1/3, 1/3, 1/12, 1/12, 1/12, 1/12], size = size),
    **data_kwargs
)
test3.plot_factors()
```
